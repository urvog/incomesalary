##Repasemos un poco los algoritmos SVM, KNN, NB

### SVM

El objetivo es encontrar un hiperplano en un espacio n-dimensional que separe los puntos de datos en sus clases potenciales. El hiperplano debe colocarse con la distancia máxima a los puntos de datos. Los puntos de datos con la distancia mínima al hiperplano se denominan vectores de soporte. Debido a su posición cercana, su influencia en la posición exacta del hiperplano es mayor que la de otros puntos de datos. En el gráfico a continuación, los vectores de soporte son los 3 puntos (2 azules, 1 verde) que se encuentran en las líneas.

```python
#One Hot Encoding para los atributos categóricos
df_income_onehot = pd.get_dummies(df_income)
```

```python
df_income_onehot
```

```python
#Dividimos nuestros datos para entrenar el modelo (Train, Validation) y para Test
df_train_ = df_income_onehot.sample(frac= .80)
df_test = df_income_onehot.drop(df_train_.index)
```

```python
#Dividimos nuesros datos train, validation and test
X = df_train_.drop('income_>50K', axis = 1)
y = df_train_['income_>50K']

X_test = df_test.drop('income_>50K', axis = 1)
y_test = df_test['income_>50K']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=0, test_size = 0.3)
```

##3.- Modelado ML

### Super Vector Machines

```python
from sklearn.svm import SVC

clf_svc = SVC(gamma = 'auto', kernel = 'sigmoid')
clf_svc.fit(X_train,y_train)
```

```python
#Imprimimos Scores
print("Train Score:", clf_svc.score(X_train,y_train))
print("Validation Score:", clf_svc.score(X_valid,y_valid))
print("Test Score:", clf_svc.score(X_test,y_test))
```

###K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier(n_neighbors = 5)
clf_knn.fit(X_train,y_train)
```

```python
#Imprimimos Scores
print("Train Score:", clf_knn.score(X_train,y_train))
print("Validation Score:", clf_knn.score(X_valid,y_valid))
print("Test Score:", clf_knn.score(X_test,y_test))
```

###Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

clf_gnb = GaussianNB(var_smoothing=1e-9)
clf_gnb.fit(X_train,y_train)
```

```python
#Imprimimos Scores
print("Train Score:", clf_gnb.score(X_train,y_train))
print("Validation Score:", clf_gnb.score(X_valid,y_valid))
print("Test Score:", clf_gnb.score(X_test,y_test))
```

##4.- Seleccionamos el Modelo

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB

clf_mnb = MultinomialNB()
clf_mnb.fit(X_train,y_train)

clf_bnb = BernoulliNB()
clf_bnb.fit(X_train,y_train)
```

##5.- Ajustamos parametors y Evaluamos

```python
#Imprimimos Scores
print("Multinomial Train Score:", clf_mnb.score(X_train,y_train))
print("Multinomial Validation Score:", clf_mnb.score(X_valid,y_valid))
print("Multinomial Test Score:", clf_mnb.score(X_test,y_test))
```

```python
#Imprimimos Scores
print("Bernoulli Train Score:", clf_bnb.score(X_train,y_train))
print("Bernoulli Validation Score:", clf_bnb.score(X_valid,y_valid))
print("Bernoulli Test Score:", clf_bnb.score(X_test,y_test))
```
